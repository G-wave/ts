# Rolling

```python
goog.rolling(window=30).mean()    # 30 business days
```

```python
rm =  goog.rolling(window='90D').mean()   # 90 Canlenda Days
rm.head()
```

```python
plt.rcParams['figure.figsize'] = [20, 10]
rolling = pd.concat([goog.Close,rm.Close], axis=1)
rolling.columns = ['Goog','Goog_90D_rolling']
rolling.plot()
```

```python
r = goog.Close.rolling('90D').agg(['mean','std'])
r.plot(subplots = True)
plt.show()
```

```python
r  = goog.Close.rolling('90D')
q1 = r.quantile(0.1)
m = r.median()
q9 = r.quantile(0.9)
```

```python
d = pd.concat([q1,m,q9], axis = 1)
d.columns = ["q1","m","q9"]
d.plot()
```

# Expanding

# Multiperiod returns

Single period return: $r_t = \frac{P_t}{P_{t-1}} - 1$ ;
Multi period return $r_t = (1+r_1)(1+r_2)...(1+r_n)-1$ 

```python
c = goog.Close.pct_change()
```

```python
c.add(1).cumprod().sub(1).mul(100).plot()
```
